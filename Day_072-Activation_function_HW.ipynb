# Rectified Linear Unit- Relu 

f(x)=max(0,x)


作業:
    寫出 ReLU & dReLU 一階導數
    並列印

```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def ReLU(x):
    l=[]
    for i in x:
        if i>0:
            l.append(i)
        else:
            l.append(0)
    return l
```

```python
def dReLU(x):
    l=[]
    for i in x:
        if i > 0:
            l.append(1)
        else:
            l.append(0)  
    return l
```

```python
def func_fig(func, x, title, r_func):
    plt.plot(x,func(x),label='linspace(-10,10,10)')
    plt.grid()
    plt.title(title)
    plt.text(4, 0.8, r_func, fontsize=15)
    plt.show()
```

```python
x=np.linspace(-10,10,100)
```

```python
func_fig(ReLU, x, 'ReLU', '$max(0,x)$')
```

```python
func_fig(dReLU, x, 'dReLU', "1 when x>0,\n0 when x <=0")
```
