```python
from sklearn import datasets, metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# 讀取鳶尾花資料集
iris = datasets.load_iris()

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25, random_state=4)

# 建立模型
clf = RandomForestClassifier()

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)
```

```python
acc = metrics.accuracy_score(y_test, y_pred)
print("Acuuracy: ", acc)
```

```python
clf_1=RandomForestClassifier(n_estimators=200)
clf_1.fit(x_train, y_train)
y_pred_1=clf_1.predict(x_test)
```

```python
acc1 = metrics.accuracy_score(y_test, y_pred_1)
print("Acuuracy: ", acc1)
```

```python
boston=datasets.load_boston()
X_train, X_test, Y_train, Y_test=train_test_split(boston.data, boston.target, random_state=20)
clr=RandomForestRegressor(random_state=20)
clr.fit(X_train, y_train)
Y_pred=clr.predict(X_test)
Y_pred_train=clr.predict(X_train)
```

```python
print("RMSE for validation:",metrics.mean_squared_error(Y_test, Y_pred)**0.5)
print("RMSE for train:", metrics.mean_squared_error(Y_train, Y_pred_train)**0.5)
print("Mean score for cross validation:",cross_val_score(clr, X_train, Y_train, cv=5).mean())
plt.plot([0,np.max(Y_test)],[0,np.max(Y_test)],linestyle='--', color='black', alpha=0.5)
plt.scatter(Y_pred, Y_test)
plt.xlabel("Predicted value")
plt.ylabel("Actaul value")
plt.title("Predicted vs Actual")
plt.show()
```

```python
#Decision Tree result
dreg=DecisionTreeRegressor(random_state=20)
dreg.fit(X_train, Y_train)
Y_pred=dreg.predict(X_test)
Y_pred_train=dreg.predict(X_train)

print("RMSE for validation:",metrics.mean_squared_error(Y_test, Y_pred)**0.5)
print("RMSE for train:", metrics.mean_squared_error(Y_train, Y_pred_train)**0.5)
print("Mean score for cross validation:",cross_val_score(dreg, X_train, Y_train, cv=5).mean())
plt.plot([0,np.max(Y_test)],[0,np.max(Y_test)],linestyle='--', color='black', alpha=0.5)
plt.scatter(Y_pred, Y_test)
plt.xlabel("Predicted value")
plt.ylabel("Actaul value")
plt.title("Predicted vs Actual")
plt.show()
```
